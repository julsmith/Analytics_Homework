```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
print(purchase_data.columns)

purchase_data.head()
purchase_data = purchase_data.set_index("Purchase ID")
purchase_data.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#### Drop duplicates 
demographics_df = purchase_data.copy()

demographics_df.drop_duplicates(subset = "SN", keep = 'first', inplace = True)

male_players = (demographics_df["Gender"]=='Male').sum()
#print(male_players)
female_players = (demographics_df["Gender"]=='Female').sum()
#print(female_players)
other_non_disclosed = (demographics_df["Gender"]=='Other / Non-Disclosed').sum()
#print(other_non_disclosed)
sum_total_players = demographics_df["Gender"].count()
#print(sum_total_players)
male_percent_players = '{:.2%}'.format((male_players/sum_total_players))
#print(male_percent_players)
female_percent_players = '{:.2%}'.format((female_players/sum_total_players))
#print(female_percent_players)
other_non_disclosed_percent = '{:.2%}'.format((other_non_disclosed/sum_total_players))
#print(other_non_disclosed_percent)
gender_demographics = pd.DataFrame([
    {"Gender": "Male", "Total Players": male_players, "Percent Players": male_percent_players},
{"Gender": "Female", "Total Players": female_players, "Percent Players": female_percent_players},
{"Gender": "Other/Non-Disclosed", "Total Players": other_non_disclosed, "Percent Players": other_non_disclosed_percent}])
gender_demographics
```

```python
unique_genders = purchase_data["Gender"].unique()
#print(unique_genders)
purchase_count = purchase_data["Gender"].count()
#print(sum_total_players)
male_players_two = (purchase_data["Gender"]=='Male').sum()
#print(male_players)
female_players_two = (purchase_data["Gender"]=='Female').sum()
#print(female_players)
other_non_disclosed_two = (purchase_data["Gender"]=='Other / Non-Disclosed').sum()
#print(other_non_disclosed)

only_males = purchase_data.loc[purchase_data["Gender"]=='Male',:]

only_females = purchase_data.loc[purchase_data["Gender"]=='Female',:]

only_other_non_disclosed = purchase_data.loc[purchase_data["Gender"]=='Other / Non-Disclosed',:]

#print(only_other_non_disclosed)

only_females_spent = only_females["Price"].sum()
only_males_spent = only_males["Price"].sum()
only_other_spent = only_other_non_disclosed["Price"].sum()

average_males = only_males_spent/male_players_two
#print(average_males)
average_females = only_females_spent/female_players_two
#print(average_females)
average_other = only_other_spent/other_non_disclosed_two
#print(average_other)
## get total Purchase value 
total_purchase_value_female = average_females*female_players_two
#print(total_purchase_value_female)
total_purchase_value_male = average_males*male_players_two
#print(total_purchase_value_male)
total_purchase_value_other = average_other*other_non_disclosed_two
#print(total_purchase_value_other)

#calculate Average total purchase per person 
average_total_purchase_female = total_purchase_value_female/female_players
#print(average_total_purchase_female)
average_total_purchase_male = total_purchase_value_male/male_players
#print(average_total_purchase_male)
average_total_purchase_other = total_purchase_value_other/other_non_disclosed
#print(average_total_purchase_other)


average_price = purchase_data["Price"].mean()
#print(average_price)

purchase_analysis = pd.DataFrame([
{"Gender": "Female",  "Average Purchase Price": average_females,"Purchase Count": female_players_two, "Total Purchase Value": total_purchase_value_female, "Average Total Purchase per Person": average_total_purchase_female },
    {"Gender": "Male", "Average Purchase Price": average_males, "Purchase Count": male_players_two, "Total Purchase Value": total_purchase_value_male, "Average Total Purchase per Person": average_total_purchase_male },
{"Gender": "Other / Non-Disclosed", "Average Purchase Price": average_other, "Purchase Count": other_non_disclosed_two,"Total Purchase Value": total_purchase_value_other, "Average Total Purchase per Person": average_total_purchase_other }])
purchase_analysis = purchase_analysis.set_index("Gender")
purchase_analysis = purchase_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Total Purchase per Person"]]
purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis["Average Total Purchase per Person"] = purchase_analysis["Average Total Purchase per Person"].map("${:.2f}".format)
purchase_analysis
```
