```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.lines as mlines
import collections

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
#city_data = city_data.drop_duplicates('city')
# Combine the data into a single dataset
combined_data = pd.merge(city_data,ride_data,how='outer', on='city')


# Display the data table for preview
combined_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#Average Fare ($) Per City
#Total Number of Rides Per City
#City Type (Urban, Suburban, Rural)

# create separate dataframes for each city type

urban_city_type = combined_data.loc[(combined_data['type'] =="Urban")]
suburban_city_type = combined_data.loc[(combined_data['type'] =="Suburban")]
rural_city_type = combined_data.loc[(combined_data['type'] =="Rural")]

#find average fare, total rides, total drivers per urban city
average_fare_urban_city = urban_city_type.groupby(['city'])['fare'].mean()
##print(average_fare_urban_city)
total_rides_urban_city = urban_city_type.groupby(['city']).count()['ride_id']
#print(total_rides_urban_city)
total_drivers_urban_city = urban_city_type.groupby(['city'])['driver_count'].mean()
#print(total_drivers_urban_city)

#find average fare, total rides, total drivers per suburban city
average_fare_suburban_city = suburban_city_type.groupby(['city'])['fare'].mean()
total_rides_suburban_city = suburban_city_type.groupby(['city']).count()['ride_id']
total_drivers_suburban_city = suburban_city_type.groupby(['city'])['driver_count'].mean()

#find average fare, total rides, total drivers per rural city
average_fare_rural_city = rural_city_type.groupby(['city'])['fare'].mean()
total_rides_rural_city = rural_city_type.groupby(['city']).count()['ride_id']
total_drivers_rural_city = rural_city_type.groupby(['city'])['driver_count'].mean()


#create plot
plt.scatter(total_rides_urban_city, average_fare_urban_city,s=total_drivers_urban_city*10, 
            marker = "o", facecolors ="lightcoral", edgecolors='black',alpha =0.5, label = "Urban")
plt.scatter(total_rides_suburban_city, average_fare_suburban_city,s=total_drivers_suburban_city*10, 
            marker = "o", facecolors ="skyblue", edgecolors='black',alpha =0.5, label = "Suburban")
plt.scatter(total_rides_rural_city, average_fare_rural_city,s=total_drivers_rural_city*10,
            marker = "o", facecolors ="gold", edgecolors='black',alpha =0.5, label = "Rural")

# title and axis labels
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
# create legend 
legend = plt.legend(title="City Type")
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]
# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
combined_data.head()
total_fares = combined_data['fare'].sum()
print(total_fares)

rural_fares = combined_data.loc[combined_data['type']=='Rural',:]
total_rural_fares = rural_fares['fare'].sum()
print(total_rural_fares)

suburban_fares = combined_data.loc[combined_data['type']=='Suburban',:]
total_suburban_fares = suburban_fares['fare'].sum()
print(total_suburban_fares)

urban_fares = combined_data.loc[combined_data['type']=='Urban',:]
total_urban_fares = urban_fares['fare'].sum()
print(total_urban_fares)

# Build Pie Chart

labels = ["Rural","Suburban","Urban"]
values = [total_rural_fares,total_suburban_fares, total_urban_fares]
colors = ["gold", "skyblue", "coral"]
explode = [0.1,0.1,0]

plt.pie(values, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")
         
# Save Figure
plt.savefig("Total Fares By City Type.png")

plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
combined_data.head()
total_rides = len(combined_data['ride_id'].unique())
#print(total_rides)

rural_rides = combined_data.loc[combined_data['type']=='Rural',:]
total_rural_rides = len(rural_rides['ride_id'])
print(total_rural_rides)

suburban_rides = combined_data.loc[combined_data['type']=='Suburban',:]
total_suburban_rides = len(suburban_rides['ride_id'])
print(total_suburban_rides)

urban_rides = combined_data.loc[combined_data['type']=='Urban',:]
total_urban_rides = len(urban_rides['ride_id'])
print(total_urban_rides)

# Build Pie Chart

labels = ["Rural","Suburban","Urban"]
values_two = [total_rural_rides, total_suburban_rides, total_urban_rides]
colors = ["gold", "skyblue", "coral"]
explode = [0.1,0.1,0]

# Build Pie Chart
plt.pie(values_two, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("Total Rides By City Type.png")
plt.show()
```
